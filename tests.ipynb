```python
import task_1 as t1
import matplotlib.pyplot as plt
```

```python
image=t1.load_image("task1_image.png")
plt.imshow(image)
plt.axis("off")
```

```python
green=t1.selective_enhancement(image,'green',blur_kernel=20,min_area=500)
yellow=t1.selective_enhancement(image,'yellow',blur_kernel=20,min_area=500)
red=t1.selective_enhancement(image,'red',blur_kernel=20,min_area=500)
```

```python
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.imshow(image)  # Original image
plt.title("Original Image")
plt.axis("off")

plt.subplot(2, 2, 2)
plt.imshow(green)
plt.title("Green Enhancement")
plt.axis("off")

plt.subplot(2, 2, 3)
plt.imshow(yellow)
plt.title("Yellow Enhancement")
plt.axis("off")

plt.subplot(2, 2, 4)
plt.imshow(red)
plt.title("Red Enhancement")
plt.axis("off")

plt.tight_layout()
plt.show()
```
